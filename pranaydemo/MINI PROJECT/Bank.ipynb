```python
class BankAccount:
    branch = "ABC"
    acc_no = 1001

    def __init__(self):
        self.name = None
        self.__password = None
        self.balance = 0.0
        self.account_no = BankAccount.acc_no
        BankAccount.acc_no += 1

    def check_balance(self):
        print(f"{self.name} ")
        print(f"Current balance: ₹{self.balance}")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{self.name}")
            print(f" Deposited ₹{amount}")
            print(f"New balance: ₹{self.balance}")
        else:
            print("Deposit amount must be greater than zero.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be greater than zero.")
        elif amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            print(f"{self.name} Amount withdrawn ₹{amount}. ")
            print(f"New balance: ₹{self.balance}")

    def change_password(self, old_pwd, new_pwd):
        if self.__password == old_pwd:
            self.__password = new_pwd
            print("Password changed successfully.")
        else:
            print("Incorrect old password.")

    @classmethod
    def change_branch(cls, new_branch):
        cls.branch = new_branch
        print(f"Branch changed to: {cls.branch}")

    @staticmethod
    def calculate_balance(credit, debit):
        return credit - debit

    @staticmethod
    def collect_user_details(account):
        print("--- Enter Account Details ---")
        account.name = input("Enter your name: ")
        account.__password = input("Set your password: ")
        initial_balance = float(input("Enter initial deposit (₹): "))
        account.balance = initial_balance

    @staticmethod
    def perform_transactions(account):
        print("--- Perform Transactions ---")

        choice = input("Do you want to deposit? (yes/no): ").lower()
        if choice == "yes":
            deposit_amount = float(input("Enter amount to deposit: ₹"))
            account.deposit(deposit_amount)

        choice = input("Do you want to withdraw? (yes/no): ").lower()
        if choice == "yes":
            withdraw_amount = float(input("Enter amount to withdraw: ₹"))
            account.withdraw(withdraw_amount)

        choice = input("Do you want to change password? (yes/no): ").lower()
        if choice == "yes":
            old_pwd = input("Enter old password: ")
            new_pwd = input("Enter new password: ")
            account.change_password(old_pwd, new_pwd)

        choice = input('Do you want to change the branch? (yes/no):').lower()
        if choice=='yes':
            new_branch=input("Enter the new branch name: ")
            account.change_branch(new_branch)

    def display_profile(self):
        print("Account Profile ")
        print(f"Account No   : {self.account_no}")
        print(f"Name         : {self.name}")
        print(f"Balance      : ₹{self.balance}")
        print(f"Bank Branch  : {self.branch}")
        print(f"Password     : {'*' * len(self.__password)} ")
        print("-------------------------")

b1=BankAccount()
BankAccount.collect_user_details(b1)
b1.display_profile()
# BankAccount.perform_transactions(b1)
# b1.display_profile()
```

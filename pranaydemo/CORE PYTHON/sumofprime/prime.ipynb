````python
class Employee:
    e_id = None
    e_role = None

    def __init__(self, name, age, sal, loc):

        if not (3 <= len(name) <= 15):
            raise ValueError("Name must be between 3 and 15 characters long.")


        if len(str(sal)) < 4:
            raise ValueError("Salary must be at least 4 digits long.")


        self.name = name
        self.age = age
        self.sal = sal
        self.loc = loc

    @classmethod
    def update_class_vars(cls, emp_id, role):
        cls.e_id = emp_id
        cls.e_role = role
        print(f"Updated Class Variables -> e_id: {cls.e_id}, e_role: {cls.e_role}")

    @staticmethod
    def get_employee_info(emp_obj):
        return {
            "Name": emp_obj.name,
            "Age": emp_obj.age,
            "Salary": emp_obj.sal,
            "Location": emp_obj.loc,
            "Employee ID": Employee.e_id,
            "Employee Role": Employee.e_role
        }

emp1 = Employee("Alice", 30, 50000, "New York")
emp2 = Employee("Bob", 28, 65000, "Chicago")
emp3 = Employee("Charlie", 35, 70000, "San Francisco")

Employee.update_class_vars(emp_id="E12345", role="Software Engineer")

print(Employee.get_employee_info(emp1))
print(Employee.get_employee_info(emp2))
print(Employee.get_employee_info(emp3))
````
